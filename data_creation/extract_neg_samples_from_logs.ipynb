```python
blocked_msg = "Your prompt has been blocked and cannot be processed at this time."

data_no_blocked = data[data['RESPONSE'] != blocked_msg]

data_no_blocked.head()
```

```python
# Unique prompt values after filtering blocked responses
unique_prompts = data_no_blocked['PROMPT'].unique()
len(unique_prompts), unique_prompts[:10]
```

```python
unique_prompts
```

```python
# DataFrame with one row per unique prompt
unique_prompts_df = pd.DataFrame(unique_prompts, columns=["PROMPT"])
unique_prompts_df.head()

# Sample ~500 rows from the filtered data (without blocked responses)
# Set random_state for reproducibility
sample_size = min(500, len(data_no_blocked))
sampled_500_df = data_no_blocked.sample(n=sample_size, random_state=42)
sampled_500_df.head()
```

```python
# Save DataFrames to disk
output_dir = "/Users/sravan/Documents/Experiments/fintuning_PII/Data/omm_PII_detection_data_dashboard"
import os
os.makedirs(output_dir, exist_ok=True)

unique_prompts_path = os.path.join(output_dir, "unique_prompts_negative_space.csv")
sampled_500_path = os.path.join(output_dir, "sampled_500.csv")

unique_prompts_df.to_csv(unique_prompts_path, index=False)
sampled_500_df.to_csv(sampled_500_path, index=False)

unique_prompts_path, sampled_500_path
```
